## Plotting

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
pi = np.pi

x = np.linspace(-4*pi, 4*pi, 1000)
plt.plot(x, np.sin(x)/x)
```

```python
#to get vectographic representation
%config InlineBackend.figure_format='svg'

plt.plot(x, np.sin(x)/x)
```

```python
plt.xticks([-4*pi,-2*pi,0,2*pi,4*pi], ['$-4\pi$','$-2\pi$','$0$','$2\pi$','$4\pi$'])
plt.yticks([0,1], ['$0$','$1$'])
plt.plot(x, np.sin(x)/x)
```

```python
plt.xticks(
    [-4 * pi, -2 * pi, 0, 2 * pi, 4 * pi],
    ['$-4\pi$', '$-2\pi$', '$0$', '$2\pi$', '$4\pi$'],
)
plt.yticks([0, 1], ['$0$', '$1$'])
plt.ylim((-0.5, 1.5))
plt.plot(x, np.sin(x) / x)
```

```python
plt.xticks(
    [-4 * pi, -2 * pi, 0, 2 * pi, 4 * pi],
    ['$-4\pi$', '$-2\pi$', '$0$', '$2\pi$', '$4\pi$'],
)
plt.yticks([0, 1], ['$0$', '$1$'])
plt.ylim((-0.5, 1.5))
plt.rcParams['figure.figsize'] = (10,4)
plt.rcParams['font.size'] = 20
plt.plot(x, np.sin(x) / x)
```

```python
plt.xticks(
    [-4 * pi, -2 * pi, 0, 2 * pi, 4 * pi],
    ['$-4\pi$', '$-2\pi$', '$0$', '$2\pi$', '$4\pi$'],
)
plt.yticks([0, 1], ['$0$', '$1$'])
plt.ylim((-0.5, 1.5))
plt.rcParams['figure.figsize'] = (10,4)
plt.rcParams['font.size'] = 20
plt.plot(x, np.sin(x) / x, label=r'$f(x)=\frac{\sin(x)}{x}$')
plt.legend(loc='best')
```
